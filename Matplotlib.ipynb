```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5, 11)
y = x **2
```

```python
x
```

```python
y
```

```python
plt.plot(x, y, 'r-')
plt.xlabel("X lable")
plt.ylabel("Y Lable")
plt.title("Title")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x, y, "r")
plt.subplot(1, 2, 2)
plt.plot(y, x, "b")
```

```python
#OO Method
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y)
axes.set_xlabel("X label")
axes.set_ylabel("Y Label")
axes.set_title("Set Title")
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])
axes1.plot(x, y)
axes1.set_title('Larger Plot')
axes2.plot(y, x)
axes2.set_title("Smaller Plot")
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
# for current_ax in axes:
#     current_ax.plot()
axes[0].plot(x,y)
axes[1].plot(y,x)
axes[1].set_title("Second Plot")
```

#### Figure Size and DPI

```python
fig = plt.figure(figsize=(3, 2), dpi=100)
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8, 2))
axes[0].plot(x, y)
axes[1].plot(y, x)
plt.tight_layout()
```

#### Save Figure

```python
fig.savefig('my_subplot.png', dpi=700)
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
ax.plot(y, x)
ax.plot(x, x**2)
ax.plot(y, y**3)
ax.set_title("Title")
ax.set_xlabel("X")
ax.set_ylabel("Y")
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x**2, label="X Square")
ax.plot(x, y**3, label="Y Cube")
ax.legend(loc=10)
```

```python
fig.savefig("plot_cube.png")
```

#### Plot Apperance
