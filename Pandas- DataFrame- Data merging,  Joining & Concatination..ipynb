```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    "A":["A1", "A2","A3", "A4", "A5"],
    "B":"B1 B2 B3 B4 B5".split(),
    "C": "C1 C2 C3 C4 C5".split(),
    "D": "D1 D2 D3 D4 D5".split()
})
df1
```

```python
df2 = pd.DataFrame({
    "A":["A6", "A7","A8", "A9", "A10"],
    "B":"B11 B12 B13 B14 B15".split(),
    "C": "C11 C12 C13 C14 C15".split(),
    "D": "D11 D12 D13 D14 D15".split()
})
df2
```

```python
df3 = pd.DataFrame({
    "A":["A10", "A20","A30", "A40", "A50"],
    "B":"B10 B20 B30 B40 B50".split(),
    "C": "C10 C20 C30 C40 C50".split(),
    "D": "D10 D20 D30 D40 D50".split()
})
df3
```

```python
left = pd.DataFrame({
    "Key1":"K1 K2 K3 K4".split(),
    "Key2":"K1 K2 K3 K4".split(),
    "A":"A1 A2 A3 A4".split(),
    "B":"B1 B2 B3 B4".split(),
    "C": "D1 D2 D3 D4".split()
})
left
```

```python
right = pd.DataFrame({
    "Key1":"K1 K2 K3 K4".split(),
    "Key2":"K1 K2 K3 K4".split(),
    "D": "D1 D2 D3 D4".split(),
    "E": "E1 E2 E3 E4".split(),
    "F": "F1 F2 F3 F4".split()
})
right
```

```python
pd.merge(left, right, on=["Key1", "Key2"] )
```

```python
pd.merge(left, right, how="inner", on=["Key1", "Key2"] )
```

```python
pd.merge(left, right, how="outer", on=["Key1", "Key2"] )
```

```python
pd.merge(left, right, how="left", on=["Key1", "Key2"] )
```

```python
pd.merge(left, right, how="right", on=["Key1", "Key2"] )
```
